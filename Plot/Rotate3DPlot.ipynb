```python
import numpy as np
import matplotlib.pyplot as plt
import math
from mpl_toolkits.mplot3d import Axes3D

#%matplotlib inline
# å¯æ—‹è½¬plot
%matplotlib notebook

# å®šä¹‰figure
fig = plt.figure()
# å°†figureå˜ä¸º3d
ax = Axes3D(fig)

# æ•°æ®æ•°ç›®
n = 10
# å®šä¹‰x, y
x = np.arange(-4, 4, 0.25)
y = np.arange(-4, 4, 0.25)

# ç”Ÿæˆç½‘æ ¼æ•°æ®
X, Y = np.meshgrid(x, y)

# è®¡ç®—æ¯ä¸ªç‚¹å¯¹çš„é•¿åº¦
R = np.sqrt(X ** 2 + Y ** 2)
# è®¡ç®—Zè½´çš„é«˜åº¦
Z = np.sin(R)

# ç»˜åˆ¶3Dæ›²é¢
ax.plot_surface(X, Y, Z, rstride = 1, cstride = 1, cmap = plt.get_cmap('rainbow'))
# ç»˜åˆ¶ä»Ž3Dæ›²é¢åˆ°åº•éƒ¨çš„æŠ•å½±
ax.contour(X, Y, Z, zdim = 'z', offset = -2, cmap = 'rainbow')

# è®¾ç½®zè½´çš„ç»´åº¦
ax.set_zlim(-2, 2)

plt.show()
```
